```python
import cv2 as cv
from matplotlib import pyplot as plt
```

### Basics

```python
def imshow(image):
    cv.imshow('image', image)
    cv.waitKey(0) 
    cv.destroyWindow('image')
```

```python
img = cv.imread("./ph/largeCat.jpg")
imshow(img)
```

```python
# resizing 
img = cv.resize(img, (640,640))
imshow(img)
```

```python

# grayscaling
grayed = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

f = plt.figure(figsize=(10,10))
f.add_subplot(121)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
f.add_subplot(122)
plt.imshow(cv.cvtColor(grayed, cv.COLOR_GRAY2RGB))
```

```python
# bluring 

blured = cv.GaussianBlur(grayed, (3,3), cv.BORDER_DEFAULT)
f = plt.figure(figsize=(10,10))
f.add_subplot(121)
plt.imshow(grayed, cmap = "gray")
f.add_subplot(122)
plt.imshow(blured, cmap = "gray")
```

```python
#canny edges
f = plt.figure(figsize=(20,20))
edges1 = cv.Canny(img, 125,175)
edges2 = cv.Canny(grayed, 125,175)
edges3 = cv.Canny(blured, 125,175)
f.add_subplot(141)
plt.imshow(edges1, cmap = "gray")
f.add_subplot(142)
plt.imshow(edges2, cmap = "gray")
f.add_subplot(143)
plt.imshow(edges3, cmap = "gray")
# заблюрим сильнее
f.add_subplot(144)
plt.imshow(cv.Canny(cv.GaussianBlur(grayed, (7,7), cv.BORDER_DEFAULT), 125,175), cmap = "gray")

```

```python
# just a glimpse on morphological operations
dilated = cv.dilate(edges3,(3,3), iterations=3)
f = plt.figure(figsize=(20,20))
f.add_subplot(121)
plt.imshow(edges3, cmap = "gray")
f.add_subplot(122)
plt.imshow(dilated, cmap = "gray")
```
